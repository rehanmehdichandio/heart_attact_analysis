```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/heart_attack_dataset.csv')
```

```python
print('Dataset Information')
df.info()
```

```python
df.head()
```

```python
df.describe(include='all')
```

```python
print('\n Print Missing values')
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Gender'] = df['Gender'].astype('category')
df['Has Diabetes'] = df['Has Diabetes'].astype('category')
df['Smoking Status'] = df['Smoking Status'].astype('category')
df['Chest Pain Type'] = df['Chest Pain Type'].astype('category')
df['Treatment'] = df['Treatment'].astype('category')
```

```python
df.dtypes
```

```python
print('\n Value Counts for Categorical Columns: ')
print(df['Gender'].value_counts())
print(df['Has Diabetes'].value_counts())
print(df['Smoking Status'].value_counts())
print(df['Chest Pain Type'].value_counts())
print(df['Treatment'].value_counts())
```

```python
#Gender Distribution
plt.figure(figsize=(8,6))
sns.countplot(x='Gender', data=df)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#Age Distribution
plt.figure(figsize=(8,6))
sns.histplot(df['Age'], kde = True, bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#Blood Pressure Distribution
plt.figure(figsize =(8, 6))
sns.histplot(df['Blood Pressure (mmHg)'], kde =True, bins = 30 )
plt.title('Blood Pressure (mmHg) Distribution')
plt.xlabel('Blood Pressure (mmHg)')
plt.ylabel('Count')
plt.show()
```

```python
#Cholesterol Dsitribution
plt.figure(figsize = (8,6))
sns.histplot(df['Cholesterol (mg/dL)'], kde=True, bins =30)
plt.title('Cholesterol Distribution')
plt.xlabel('Cholesterol')
plt.ylabel('Counts')
plt.show()
```
